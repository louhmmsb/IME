```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Cronômetros

```python
def temposMedios(matTravessia):
    return(np.mean(matTravessia[:,1:], axis = 0))

def analisaDados(matTravessia, temposMediosMovimento):

    print('Tempos medios: \n',temposMediosMovimento)

    # desvio medio quadratico em relacao a media

    print('\n\n===Desvios quadraticos em relacao a media===\n')
    matrizMedia = np.repeat([temposMediosMovimento],4, axis=0)
    matrizDiferenca = matTravessia[:,1:] - matrizMedia
    matrizDiferenca = np.power(matrizDiferenca,2)
    desvios = np.mean(matrizDiferenca,axis=0)

    print('Erros quadráticos: \n',matrizDiferenca)
    print('\n Desvios: \n',desvios)

    posicoes = cronometros[0,1:]
    fig, ax = plt.subplots()
    plt.errorbar(posicoes,temposMediosMovimento,yerr=10*desvios)
    plt.title('Dados dos cronometros da travessia')
    ax.set_xlabel('Posicao (metros)')
    ax.set_ylabel('Tempo (segundos)')
    plt.show()
    
```

```python
# leitura dos dados

cronometros = pd.read_csv("https://www.ime.usp.br/~cesar/courses/mac0209/travessia_cronometros.csv")

cronometros = np.array(cronometros)

print('Dados originais:\n')
print(cronometros)

################# MOVIMENTO UNIFORME
print('\n\n Movimento Uniforme\n')

dadosMovimentoUniforme = cronometros[np.where(cronometros[:,0] == 0)]

temposMediosMovimento = temposMedios(dadosMovimentoUniforme)
analisaDados(dadosMovimentoUniforme,temposMediosMovimento)

################# MOVIMENTO ACELERADO

print('\n\n Movimento Acelerado\n')

dadosMovimentoAcelerado = cronometros[np.where(cronometros[:,0] == 1)]

temposMediosMovimento = temposMedios(dadosMovimentoAcelerado)
analisaDados(dadosMovimentoAcelerado,temposMediosMovimento)

################# COMPARACAO DOS MOVIMENTOS UNIFORME E ACELERADO

print('\n\n Comparacao dos Movimentos Uniforme e Acelerado\n')

temposMediosMovimento = temposMedios(dadosMovimentoUniforme)
analisaDados(dadosMovimentoAcelerado,temposMediosMovimento)
```
